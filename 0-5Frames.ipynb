```python
import tensorflow
from tensorflow import keras
import numpy
tensorflow.config.list_physical_devices('GPU')
```

```python

inputLayer = keras.Input(shape=(10,))
hidden1 = keras.layers.Dense(8, activation="relu")(inputLayer)
hidden2 = keras.layers.Dense(4, activation="relu")(hidden1)
outputLayer = keras.layers.Dense(1, activation="sigmoid")(hidden2)
model = keras.Model(inputs=inputLayer, outputs=outputLayer)
model.summary()
```

```python
import csv
f = open("scoreCardAll.csv")
dataReader = csv.reader(f)
next(dataReader)
dataArray = []
for row in dataReader:
    if len(row) != 0:
        dataArray.append(row)
f.close()
```

```python
inputArray = []
outputArray = []
for row in dataArray:
    for x in range(201):
        row[x] = float(row[x])
    row[200] = row[200]/300
    inputArray.append(row[:10])
    outputArray.append(row[-1:])
inputModelArray = numpy.array(inputArray[:-500])
outputModelArray = numpy.array(outputArray[:-500])
inputTestArray = numpy.array(inputArray[-500:])
outputTestArray = numpy.array(outputArray[-500:])
```

```python
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"]
)
```

```python
model.fit(inputModelArray, outputModelArray, batch_size=16, epochs=10, validation_split=.2)
```

```python
numCorrect = 0
for x in range(500):
    guess = int(model.predict(inputTestArray[x].reshape(1, 10))[0][0]*300)
    actual = int(outputTestArray[x][0]*300)
    if abs(guess - actual) < 15:
        numCorrect += 1
```

```python
print(int(model.predict(inputTestArray[13].reshape(1, 10))[0][0]*300))
print(int(outputTestArray[13][0]*300))
print(numCorrect)
```
